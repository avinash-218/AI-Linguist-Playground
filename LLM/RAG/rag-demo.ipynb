```python
from langchain.vectorstores import Chroma
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.chains import RetrievalQA
from langchain.document_loaders import TextLoader, DirectoryLoader

import transformers
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
```

```python
!wget https://github.com/Shafi2016/Youtube/raw/main/stock_market_june_2023.zip -O stock_market_june_2023.zip
!unzip stock_market_june_2023.zip
```

```python
loader = DirectoryLoader('/kaggle/working/stock_market_june_2023', glob='./*.txt', loader_cls=TextLoader)

documents = loader.load()
```

```python
#documents
len(documents)
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
texts = text_splitter.split_documents(documents)
```

```python
len(texts)
#texts[0]
```

```python
model = 'tiiuae/falcon-7b-instruct'

tokenizer = AutoTokenizer.from_pretrained(model)

pipeline = transformers.pipeline(
    'text-generation',
    model=model,
    tokenizer=tokenizer,
    device_map='cuda'
)
```
